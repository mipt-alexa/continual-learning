```python
def plot_full(run_name):
    if run_name[-5:] != ".json":
        run_name += ".json"
    with open(f'./pipeline/results/{run_name}', 'r') as f:
        data = json.load(f)

    fig, ax = plt.subplots(1, 2, figsize=(8,2))

    ax[0].plot(data["loss"], label="loss")
    ax[1].plot(data["acc_train"], label="train acc")
    ax[1].plot(data["acc_val"], label="val acc")

    plt.title(run_name)
    ax[0].legend()
    ax[1].legend()
    plt.show()
```

```python
run_names = ["5.00e-03_adam_50_resnet",
             "5.00e-04_adam_50_resnet",
             "5.00e-05_adam_50_resnet",
             "5.00e-03_adam_50_vit",
            "5.00e-04_adam_50_vit",
            "5.00e-05_adam_50_vit"]

for name in run_names:
    plot_full(name)
```

```python
def plot_cil(run_name):
    if run_name[-5:] != ".json":
        run_name += ".json"
    with open(f'./pipeline/results/{run_name}', 'r') as f:
      data = json.load(f)
    data.keys()

    plt.figure(figsize=(6,3))
    plt.plot(data["acc_last"], label="val acc on last task")
    plt.plot(data["num_epochs"]*np.arange(1, 10 + 1), data["acc_0"], 'o', label="val acc on 0 task")
    plt.title(run_name)
    plt.legend()
    plt.show()
```

```python
run_name = ["5.00e-04_adam_25_resnet_cil", 
            "5.00e-05_adam_25_resnet_cil",
            "vit_cil_adam_1.00e-05_25",
            "vit_cil_adam_5.00e-05_25"]

for name in run_name:
    plot_cil(name)
```

## LR Scheduler visualizer

```python
model = torch.nn.Linear(10, 1)
optimizer = optim.Adam(model.parameters(), lr=1e-3)

n = 40
lr_vals = np.zeros(n+1)

scheduler_1 = optim.lr_scheduler.ConstantLR(optimizer, factor=1)
scheduler_2 = optim.lr_scheduler.LinearLR(optimizer, start_factor=0.1, end_factor=1, total_iters=5)
scheduler_3 = optim.lr_scheduler.ExponentialLR(optimizer, gamma=0.97)
result_scheduler = optim.lr_scheduler.SequentialLR(optimizer, [scheduler_2, scheduler_1, scheduler_3], milestones=[5, 10])

for i in range(n+1):
    lr_vals[i] = optimizer.param_groups[0]["lr"]    
    result_scheduler.step()


plt.plot(np.arange(n+1), lr_vals)
```
